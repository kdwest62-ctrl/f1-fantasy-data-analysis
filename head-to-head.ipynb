```python
import matplotlib.pyplot as plt

driver1_name = input("Driver 1: ")
driver1_pts = int(input("Total points: "))
driver2_name = input("Driver 2: ")
driver2_pts = int(input("Total points: "))

values = [driver1_pts, driver2_pts]
categories = [driver1_name, driver2_name]
plt.pie(values, labels=categories)
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def stand_dev(nums):
  avg = np.mean(nums)
  sqrd_diff = []
  for num in nums:
    sqrd_diff.append((avg - num)**2)
  result = math.sqrt(sum(sqrd_diff)/len(sqrd_diff))
  return round(result, 2)

driver1 = input('Driver 1: ')
driver1_pts = input('Fantasy points progression (use spaces): ').split()
dataset1 = np.array(list(map(lambda x: int(x), driver1_pts)))
result1 = stand_dev(dataset1)

driver2 = input('Driver 2: ')
driver2_pts = input('Fantasy points progression (use spaces): ').split()
dataset2 = np.array(list(map(lambda x: int(x), driver2_pts)))
result2 = stand_dev(dataset2)

x_axis = [driver1, driver2]
y_axis = [result1, result2]

plt.bar(x_axis, y_axis)
plt.xlabel('Driver')
plt.ylabel('Standard Deviation')
plt.show()
```
