```python
import numpy as np

def mean(nums):
  avg = np.mean(nums)
  return round(avg, 2)

points = input('Fantasy points progression: ').split() # separate using spaces
data = np.array(list(map(lambda x: int(x), points)))
print(f"Average points: {mean(data)}")
```

```python
import math
import numpy as np

def stand_dev(nums):
  avg = np.mean(nums)
  sqrd_diff = []
  for num in nums:
    sqrd_diff.append((avg - num)**2)
  result = math.sqrt(sum(sqrd_diff)/len(sqrd_diff))
  return round(result, 2)

points = input('Fantasy points progression: ').split() # separate using spaces
data = np.array(list(map(lambda x: int(x), points)))
print(f"Standard deviation: {stand_dev(data)}")
```
