```python
import matplotlib.pyplot as plt

driver1 = input('Driver: ')
dataset1 = input("Points progression: ").split() # separate using spaces
x_axis = list(map(lambda n: int(n), dataset1))
driver2 = input('Driver: ')
dataset2 = input("Points progression: ").split() # separate using spaces
y_axis = list(map(lambda n: int(n), dataset2))
plt.scatter(x_axis, y_axis)
plt.xlabel(driver1)
plt.ylabel(driver2)
plt.grid()
plt.show()
```
